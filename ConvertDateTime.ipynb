```python
import pandas as pd
import datetime as dt
from sqlalchemy import create_engine
```

```python
# Pull in meteorite csv data
meteorite_df = pd.read_csv("Resources/Cleaned_Meteorite_Landings.csv")
meteorite_df
```

```python
# convert year column to datetime
meteorite_df['year'] = pd.to_datetime(meteorite_df['year'], format='%m/%d/%Y')
meteorite_df
```

```python
# Convert year column to show just the year
meteorite_df['year'] = meteorite_df['year'].dt.year
meteorite_df
```

```python
# save to CSV
meteorite_df.to_csv("Resources/Meteorite_Landings_Years.csv")
```

```python
# Pull in fireball csv data
fireball_df = pd.read_csv("Resources/Fireball_Data_CNEOS.csv")
fireball_df
```

```python
# convert year column to datetime
fireball_df['Peak Brightness Date/Time (UT)'] = pd.to_datetime(fireball_df['Peak Brightness Date/Time (UT)'], format='%m/%d/%Y %H:%M')
fireball_df
```

```python
# Convert year column to show just the year
fireball_df['Peak Brightness Date/Time (UT)'] = fireball_df['Peak Brightness Date/Time (UT)'].dt.year
fireball_df
```

```python
# save to CSV
fireball_df.to_csv("Resources/Fireball_Data_Years.csv")
```
